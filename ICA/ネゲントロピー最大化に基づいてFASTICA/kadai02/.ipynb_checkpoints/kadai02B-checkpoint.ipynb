```python
import random
from numpy import *
import matplotlib.pyplot as plt
from scipy.io import wavfile
import copy

def load_data():
    fs1, data1 = wavfile.read('speechB1.wav')
    fs2, data2 = wavfile.read('speechB2.wav')
    fs3, data3 = wavfile.read('speechB3.wav')
    X1=data1-mean(data1)
    X2=data2-mean(data2)
    X3=data3-mean(data3)
    X=array([X1,X2,X3],dtype=float)    
    return X,fs1;

def whiten(X):
    xx=dot(X,X.T)
    D,E = linalg.eig(xx)
    D=diag(1/sqrt(D))
    V = dot(D, E.T)
    Z=dot(V,X)
    return Z



def Decorrelation(W):
    d, e = linalg.eigh(dot(W, W.T))
    return dot(dot(e * (1. / sqrt(d)), e.T), W)

def do_fastica(Z):
    n, m = Z.shape;
    Z *= sqrt(Z.shape[1])
    W = ones((n,n), float32)
    y=ones((n,m), float32)
    for i in range(n): 
        for j in range(n):
            W[i][j] = random.random()
    for i in range(n):  
        W[i]= W[i]/linalg.norm(W[i])
        while True:
            W1=copy.copy(W)
            y[i] = dot(W[i].T,Z)
            for j in range(n):
                W[i][j] = mean((Z*(y[i]**3))[j]) - 3*W[i][j]
            W[i] = W[i]/linalg.norm(W[i])
            W = Decorrelation(W)
            lim = abs(abs(dot(W[i].T,W1[i]))-1)
            if lim < 0.000000000001:
                break
    return W



X,fs=load_data()
Z=whiten(X)
W = do_fastica(Z)
S = dot(W,Z)
wavfile.write('kadai02B_01.wav',fs,S[0])
wavfile.write('kadai02B_02.wav',fs,S[1])
wavfile.write('kadai02B_03.wav',fs,S[2])
```
